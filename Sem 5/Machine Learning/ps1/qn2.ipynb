```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5,5)
```

```python
fig, asx = plt.subplots(1, 2)
y = np.log(x)
asx[0].plot(x, y)
asx[0].set_title("log x")

asx[1].plot(x, -y)
asx[1].set_title("- log x")

plt.show()
```

```python
fig, asx = plt.subplots(1, 2)
y = np.multiply(x, np.log(x))
asx[0].plot(x, y)
asx[0].set_title("x * log x")

asx[1].plot(x, -y)
asx[1].set_title("- (x * log x)")
plt.show()
```

```python
fig, asx = plt.subplots(1, 2)
y = np.multiply(x, np.log2(x))
asx[0].plot(x, y)
asx[0].set_title("x * log2 x")

asx[1].plot(x, -y)
asx[1].set_title("- (x * log2 x)")
plt.show()
```

```python
fig, asx = plt.subplots(1, 2)
y = np.multiply(x, np.log10(x))
asx[0].plot(x, y)
asx[0].set_title("x * log10 x")

asx[1].plot(x, -y)
asx[1].set_title("- (x * log10 x)")
plt.show()
```

```python
fig, asx = plt.subplots(1, 2)
y = np.exp(x)
asx[0].plot(x, y)
asx[0].set_title("exp x")

asx[1].plot(x, -y)
asx[1].set_title("- exp x")
plt.show() 
```

```python
fig, asx = plt.subplots(1, 2)
y = np.power(x, 2)
asx[0].plot(x, y)
asx[0].set_title("x^2")

asx[1].plot(x, -y)
asx[1].set_title("- x^2")

plt.show()
```

```python
fig, asx = plt.subplots(1, 2)
y = 1/np.power(x, 5)
asx[0].plot(x, y)
asx[0].set_title("1/(x^5)")

asx[1].plot(x, -y)
asx[1].set_title("- 1/(x^5)")

plt.show()
```

```python
fig, asx = plt.subplots(1, 2)
y = np.abs(x)
asx[0].plot(x, y)
asx[0].set_title("|x|")

asx[1].plot(x, -y)
asx[1].set_title("- |x|")

plt.show()
```

```python
fig, asx = plt.subplots(1, 2)
print("Convex + Concave = Convex")
y = np.exp(x) - np.power(x, 2)
asx[0].plot(x, y)
asx[0].set_title("exp x - x^2")

asx[1].plot(x, -y)
asx[1].set_title("- (exp x - x^2)")
plt.show() 
```

```python
fig, asx = plt.subplots(1, 2)
print("Convex + Concave = Convex")
y = np.exp(x) + np.abs(x)
asx[0].plot(x, y)
asx[0].set_title("exp x - |x|")

asx[1].plot(x, -y)
asx[1].set_title("- (exp x - |x|)")
plt.show() 
```
